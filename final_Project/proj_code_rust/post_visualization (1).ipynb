```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the clustered_sectors.csv file
clustered_sectors_df = pd.read_csv("clustered_sectors2.csv")

# Create a contingency table to count occurrences of each Sector in each cluster
contingency_table = pd.crosstab(clustered_sectors_df['cluster'], clustered_sectors_df['sector'])

# Visualize the contingency table using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Cluster vs Sector Composition')
plt.xlabel('Sector')
plt.ylabel('Cluster')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
